```python
#regression
from sklearn  import linear_model 
import numpy as np
```

```python
df_train_p= df_pop_year[['Area Name','Total population','Year']]
categorical_columns = ['Area Name']
df_encode_p = pd.get_dummies(data = df_train_p, prefix = 'OHE', prefix_sep='_',
               columns = categorical_columns,
              dtype='int8')
X_train_p= df_encode_p.drop('Total population', axis=1)
y_train_p= df_encode_p['Total population']
```

```python
X_test= df_train_p[df_train_p['Year']==1991]
X_test['Year']=2021
X_test=X_test.drop(columns='Total population')
X_test=X_test.append(X_test, ignore_index=True)
X_test.loc[33:,'Year']= 2031
X_test=X_test.append(X_test[X_test['Year']== 2031], ignore_index=True)
X_test.loc[66:,'Year']= 2041
X_test_p= X_test.copy()
```

```python
categorical_columns = ['Area Name']
X_test_encode = pd.get_dummies(data = X_test, prefix = 'OHE', prefix_sep='_',
               columns = categorical_columns, 
              dtype='int8')
```

```python
lin_reg = linear_model.LinearRegression()
lin_reg.fit(X_train_p,y_train_p)
```

```python
y_pred=lin_reg.predict(X_test_encode)
```

```python
X_test_p['Total population']=y_pred
```

```python
df_pop_final= df_train_p.append(X_test_p, ignore_index=True)
df_pop_final.head() 
```
